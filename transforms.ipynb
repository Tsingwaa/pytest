```python
import torch
from PIL import Image
from torchvision import transforms as T

img_path = './lena.jpg'
```

```python
PIL_img = Image.open(img_path).convert('RGB')
```

```python
img = T.ToTensor()(PIL_img)
```

```python
noise = 0.01 * torch.randn_like(img)
noised_img = img + noise 
```
